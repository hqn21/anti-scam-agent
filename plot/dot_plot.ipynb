```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('websites.csv')
```

### Duration

```python
fig, ax = plt.subplots(figsize=(10, 6))
avg_duration = df['Duration'].mean()

ax.scatter(df.index, df['Duration'], color='royalblue', alpha=0.7, edgecolors='none', s=50, label='Sample Duration')
ax.axhline(avg_duration, color='salmon', linestyle='--', label=f'Average duration: {avg_duration:.1f}s')

ax.set_xlabel('Website Index', fontsize=12)
ax.set_ylabel('Duration (s)', fontsize=12)
ax.set_title('Duration per Website', fontsize=16, fontweight='bold')
ax.legend(loc='upper right', fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('Duration per Website.png', dpi=900)
plt.show()
```

### Cost

```python
fig, ax = plt.subplots(figsize=(10, 6))
avg_cost = df['Cost'].mean()

ax.scatter(df.index, df['Cost'], color='olivedrab', alpha=0.7, edgecolors='none', s=50, label='Sample Cost')
ax.axhline(avg_cost, color='darkgray', linestyle='--', label=f'Average Cost: ${avg_cost:.3f}')

ax.set_xlabel('Website Index', fontsize=12)
ax.set_ylabel('Cost (USD)', fontsize=12)
ax.set_title('Cost per Website', fontsize=16, fontweight='bold')
ax.legend(loc='upper right', fontsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('Cost per Website.png', dpi=900)
plt.show()
```
